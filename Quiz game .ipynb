```python
#Quiz Using the python____ 
questions =("1.Who is the father of Python programming.?", 
          "2.Symbol used in python to comment a line.?",  
          "3.Number of KeyWord are there in python.?", 
          "4.Which of the following datatype is immutable.?",
          "5.Which datatype doesn't allow the duplicates.?")

# options for the quations

options =(("A.Dennis Ritchie ","B.James Gosling ","C.Guido Van Rossum ","D.Bjarne Stroustrup "),
          ("A.%","B.#","C.@","D.&"),
          ("A.32 ","B.33","C.34","D.35"),
          ("A.List","B.Tuple","C.Sets","D.Dictionary "),
          ("A.List","B.Tuple","C.Sets","D.Dictionary "))

# Answers for the questions

answers =("C","B","D","B","C")          
guesses =[]
score =0
questions_num =0

# For looop for printing the Questions
print("*-------------------- Start Quiz---------------*")

for question in questions:
        print("-----------------------------")
        print(question)
        for option in options[questions_num]:
               print(option)
        guess =input("Enter (A, B, C, D):").upper()
        guesses.append(guess) 
        if guess ==answers[questions_num]:
            score +=1
            print("CORRECT ANSWER!")
        else:
            print("INCORRECT ANSWER")
            print(f"{answers[questions_num]} is the correct")
        questions_num +=1
# Results

print("*****************") 
print("      RESULTS    ")
print("*****************") 

# For gettig the results 
print("answers:",end=" ") 
for answer in answers:
    print(answer,end=" ")
print()

# For printing the guesses 

print("guesses:",end=" ")
for guess in guesses:
    print(guess,end=" ")
print()      

# For calculation the total score 

score = int(score/len(questions)*100)
print(f"Your score is:{score}")

#Feed back about the Quizzer 
if score >= 80 and score <= 100:
        print("Your performance is Excellent in python programming")
elif score >= 70 and score <= 80:
        print("Your performance is Good in python programming ")
elif score >= 60 and score <= 70:
        print("Your performance is Satisfactory in python programming ")
else:
        print("Your performance is Average in python programming prepare well for next time!")


```
